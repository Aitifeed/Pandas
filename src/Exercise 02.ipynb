```python
import pandas as pd

df = pd.read_csv(
    "C:\\Users\\Lenovo\\Desktop\\Тест\\Занятие по pandas\\1 занятие\\pandas\\data\\auto.csv",
    index_col = 'ID')
```

```python
df.count()
```

```python
df.drop_duplicates(['CarNumber','Make_n_model', 'Fines'],inplace = True)
```

```python
df.sort_values(['CarNumber','Make_n_model','Fines'],ascending=True,inplace=True)
```

```python
df.drop_duplicates(['CarNumber','Make_n_model','Fines'],keep='last',inplace= True)
```

```python
#Проверка на дубликаты
df.loc[df[['CarNumber','Make_n_model','Fines']].duplicated()]
```

```python
df.isnull().sum()
```

```python
df.dropna(thresh=500,inplace=True,axis=1)
```

```python
df.isna().sum()
```

```python
df['Refund']=df['Refund'].ffill()
```

```python
avg = df['Fines'].dropna().mean()
df['Fines'] = df['Fines'].fillna(value = avg)
```

```python
df['Make'] = df['Make_n_model'].apply(lambda x:x.split()[0])
df['Model'] = df['Make_n_model'].apply(
    lambda x:x.split()[1] if len(x.split())>1 else None)
```

```python
df.isnull().sum()
```

```python
df.drop('Make_n_model',axis=1,inplace=True)
```

```python
df.to_json('auto.json',orient='records')
```
