```python
#Убираем наши индекс и переименовываем его обратно
df.reset_index(inplace = True)
df.rename(columns={'date_time':'datetime'},inplace = True)
```

```python
df['datetime'] = df['datetime'].astype('datetime64[ns]')
```

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df['second'] = df['datetime'].dt.second
```

```python
df.set_index('user',inplace = True)
```

```python
df['hh.mm'] = df['datetime'].dt.strftime('%H.%M')
df['hh.mm'] = df['hh.mm'].astype('float')
```

```python
bins = [0, 3.59, 6.59, 10.59, 16.59, 19.59, 23.59]
labels = ['night','early morning','morning','afternoon','early evening','evening']
df['daytime']=pd.cut(df['hh.mm'],bins = bins,labels = labels)
```
